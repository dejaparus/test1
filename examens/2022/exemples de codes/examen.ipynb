```python
#exercice 1 
import numpy as np 
import matplotlib.pyplot as plt 

x = np.linspace(-10,10,50)

cout = lambda a,x : np.exp(a*x) - a * x -1

def plot_cout(a):
    
    plt.plot(x, cout(a,x))
    plt.show()
    

p1 = plot_cout(0.1) 
p2 = plot_cout(0.5) 
p3 = plot_cout(1.2) 
     
```
